```python
import tkinter as tk
from tkinter import ttk
from PIL import ImageTk,Image
from pytube import YouTube
import os
import time
```

```python
class window():
    # function to reset the screen
    def on_reset(self):
        self.__init__(root)
    
    
    def progress_start(self,stream = None, chunk = None, file_handle = None, remaining = None):
        self.pb.start()
    def progress_stop(self,stream = None, chunk = None, file_handle = None, remaining = None):
        self.pb.stop()
    
    
    #Function to update the dropdowns for type and quality
    def dropdown_update(self):
        global typeslist,qualitylist
        strm=self.yt.streams.filter(progressive=True)
        for item in strm:
            if item.mime_type[6:] not in typeslist:
                typeslist.append(item.mime_type[6:])
        for item in strm:
            if item.resolution not in qualitylist:
                qualitylist.append(item.resolution)

        self.type_list = tk.OptionMenu(self.frame2,self.types,*typeslist)
        self.type_list.place(x=190,y=100)
        self.quality_list = tk.OptionMenu(self.frame2,self.qualities,*qualitylist)
        self.quality_list.place(x=450,y=100)
        
    #function to fetch url stream data 
    def on_fetch(self):
        try:
            url = self.url_entry.get()
            if url[0:5]=="https":
                url=url[5:]
            self.yt = YouTube(url,on_progress_callback=self.progress_start,on_complete_callback=self.progress_stop)
            self.fetch_message.set(self.yt.title)
            self.dropdown_update()
        except:
            self.fetch_message.set(self.failure_case.get())
        
    
    #Function to execute download        
    def on_download(self):
        formt=self.types.get()
        resolution = self.qualities.get()
        self.video= self.yt.streams.filter(progressive = True, file_extension = formt,resolution = resolution).first()
        self.video.download("C:\\Users\\abhik\\Downloads",skip_existing= False) 
        self.label5.config(text = "Congrats your video was downloaded succesfully.")

  
    def __init__(self,root):
        img1 = ImageTk.PhotoImage(Image.open("youtube.png"))
        img2 = ImageTk.PhotoImage(Image.open("download.png").resize((200,80)))
        img3 = ImageTk.PhotoImage(Image.open("fetch.png").resize((200,70)))
        img4 = ImageTk.PhotoImage(Image.open("reset.png").resize((120,60)))
        global typeslist,qualitylist
        typeslist=['']
        qualitylist = ['']
        self.sucess_case = tk.StringVar(root,"URL fetched successfully, Kindly select the format and quality")
        self.fetch_message = tk.StringVar(root," ")
        self.failure_case = tk.StringVar(root,"URL was not fetched properly,Kindly cross check\n(make sure that it is a Youtube URL)")
        self.types = tk.StringVar(root,typeslist[0])   #default value for types
        self.qualities = tk.StringVar(root,qualitylist[0]) #default value for quality
        
        
    
        #frame 1 includes logo, URL entry field and fetch button
        self.frame1 = tk.Frame(root)
        self.frame1.place(x=1,y=1,height = 300,width = 700)   
        
        #label for logo
        self.logo_label = tk.Label(self.frame1,image=img1)
        self.logo_label.image=img1
        self.logo_label.place(x=10,y=10,height = 100,width = 700)
        
        #Entry field for URL
        self.url_entry = tk.Entry(self.frame1)
        self.url_entry.place(x=200,y=120,height = 30,width = 300)
        
        #Fetch button
        self.fetch_btn = tk.Button(self.frame1,image=img3,borderwidth=0,command=self.on_fetch)
        self.fetch_btn.image=img3
        self.fetch_btn.place(x=250,y=180)
        
        #frame 2 includes success or failure message, dropdown box for quality and type
        self.frame2 = tk.Frame(root)
        self.frame2.place(x=1,y=280,height = 300,width = 700)
        
        #label for success or error message
        self.validlabel = tk.Label(self.frame2,textvariable=self.fetch_message,font=("ComiC Sans MS",12),width =50)
        self.validlabel.place(x=100,y=0)  
        
        #label for Type of stream
        self.label3 = tk.Label(self.frame2,text="Type",font=("ComiC Sans MS",12))
        self.label3.place(x=200,y=70)  
        
        #label for quality of stream
        self.label4 = tk.Label(self.frame2,text="Quality",font=("ComiC Sans MS",12))
        self.label4.place(x=450,y=70)  
        
        
        #listbox for type of stream
        self.type_list = tk.OptionMenu(self.frame2,self.types,*typeslist)
        self.type_list.place(x=190,y=100)
        
        #Listbox foo quality of stream
        self.quality_list = tk.OptionMenu(self.frame2,self.qualities,*qualitylist)
        self.quality_list.place(x=450,y=100)
        
        
        
        #Frame 3 for download button, progress bar and final message
        self.frame3 = tk.Frame(root)
        self.frame3.place(x=1,y=450,height = 300,width = 700)   
        
        # Download button
        self.download_btn = tk.Button(self.frame3,image=img2,borderwidth=0,command = self.on_download)
        self.download_btn.image=img2
        self.download_btn.place(x=250,y=0)
        
        #progress bar
        self.pb = ttk.Progressbar(self.frame3, orient='horizontal',length=200, mode='indeterminate')
        self.pb.place(x=250,y=100)
        
        self.label5 = tk.Label(self.frame3,text=" ",font=("ComiC Sans MS",12),width=60,wraplength=550)
        self.label5.place(x=50,y=150)  

        self.reset_btn = tk.Button(self.frame3,image=img4,borderwidth=0,command = self.on_reset)
        self.reset_btn.image=img4
        self.reset_btn.place(x=285,y=220)
    
```

```python
global img1
root = tk.Tk()
root.geometry('700x800')
root.title('Youtube Downloader')
win1=window(root)

root.mainloop()
```
